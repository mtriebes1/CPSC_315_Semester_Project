### Min Temp & Rainfall Relationship


```python

table = MyPyTable()

table.load_from_file("Sydney_weather.csv")

x = table.get_column("MinTemp", False)
x_float = []
y = table.get_column("Rainfall", False)
y_float = []

for i in range(len(x)):
    if(x[i] != 'NA' and y[i] != 'NA'):
        x_float.append(float(x[i]))
        y_float.append(float(y[i]))

labels = ["Min Temp & Rainfall Relationship", "Min Temp (Degrees C)", "Rain Fall (cm)"]

utils.create_scatter_plot(x_float, y_float, labels)
```

### How Often Does it Rain? 

```python
rain_data = table.get_column("RainToday", False)

did_rain = 0
didnt_rain = 0
counts = []

for i in range(len(rain_data)):
    if (rain_data[i] != 'NA'):
        if (rain_data[i] == 'No'):
            didnt_rain += 1
        elif (rain_data[i] == 'Yes'):
            did_rain += 1

counts.append(did_rain)
counts.append(didnt_rain)

utils.create_pie_chart(['Yes', 'No'], counts, "Did it Rain Today?")
```

### Min Temperature & Rainfall When It Actually Rains

```python


x = table.get_column("MinTemp", False)
y = table.get_column("Rainfall", False)

rainfall = []
temps = []

for i in range(len(rain_data)):
    if (rain_data[i] != 'NA'):
        if (rain_data[i] == 'Yes'):
            if (x[i] != 'NA' and y[i] != 'NA'):
                rainfall.append(y[i])
                temps.append(x[i])

labels = ["Min Temp & Rainfall Relationship (w/ Rain)", "Min Temp (Degrees C)", "Rain Fall (cm)"]
utils.create_scatter_plot(temps, rainfall, labels)

```
