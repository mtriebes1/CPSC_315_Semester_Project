### Imports

```python
import utils
from mysklearn.mypytable import MyPyTable
```

### Min Temp & Rainfall Relationship


```python

table = MyPyTable()

table.load_from_file("Sydney_weather.csv")

x = table.get_column("MinTemp", False)
x_float = []
y = table.get_column("Rainfall", False)
y_float = []

for i in range(len(x)):
    if(x[i] != 'NA' and y[i] != 'NA'):
        x_float.append(float(x[i]))
        y_float.append(float(y[i]))

labels = ["Min Temp & Rainfall Relationship", "Min Temp", "Rain Fall"]

utils.create_scatter_plot(x_float, y_float, labels)
```
